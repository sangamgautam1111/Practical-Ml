```python
import pandas as pd
data = pd.read_csv('housing.csv')
df = pd.DataFrame(data)
#splitting train set
from sklearn.model_selection import train_test_split
train_set , test_set = train_test_split(df , test_size = 0.2 , random_state= 42)
#clearning the train data
from sklearn.impute import SimpleImputer
imputer = SimpleImputer(strategy="median")
housing_num = train_set.drop("ocean_proximity" , axis=1)
X =imputer.fit_transform(housing_num)
housing_tr = pd.DataFrame(X , columns=housing_num.columns , index= housing_num.index)
housing_tr["ocean_proximity"] = train_set["ocean_proximity"]
housing_tr
```

```python
categorical = housing_tr[["ocean_proximity"]]
#handaling categorical data using ordinal encoding
# from sklearn.preprocessing import OrdinalEncoder
# ordinal_encoder  = OrdinalEncoder()
# housing_cat_encoded = ordinal_encoder.fit_transform(categorical)
# housing_cat_encoded = pd.DataFrame(housing_cat_encoded , columns=["ocean_proximity"] , index=housing_tr.index)
# housing_tr["ocean_proximity"] = housing_cat_encoded
# housing_tr
#using one hot encoding
from sklearn.preprocessing import OneHotEncoder
one_hot_encoder  = OneHotEncoder()
housing_cat_1hot = one_hot_encoder.fit_transform(categorical)
housing_cat_1hot = housing_cat_1hot.toarray()
housing_tr["ocean_proximity"] = housing_cat_1hot
```
