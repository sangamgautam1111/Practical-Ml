```python
import pandas as pd
data = pd.read_csv("paddydataset.csv")
data_frame = pd.DataFrame(data)
```

```python
data_frame.head()
```

```python
#from the data splitting train test
from sklearn.model_selection import train_test_split
train_ , test_ = train_test_split(data_frame  , test_size=0.2 , random_state=42)
#feature and label sepration from the train dataset
X = train_.drop('Paddy yield(in Kg)' , axis = 1)
Y = train_['Paddy yield(in Kg)']
```

```python
#visualizing the data
import matplotlib.pyplot as plt
plt.scatter(X['30DRain( in mm)'] , Y)
plt.xlabel("Rainfall (30 days)")
plt.ylabel("paddy yeild in kg")
plt.show()
```

```python
#visualizing the data by scatter matrix plot
from pandas.plotting import scatter_matrix
scatter_matrix(X , figsize=(10,11))
plt.show()
```
