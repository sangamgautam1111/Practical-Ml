```python
import pandas as pd
data = pd.read_csv("housing.csv")
data_frame = pd.DataFrame(data)
#spliting train and test set
from sklearn.model_selection import train_test_split
train_set , test_set = train_test_split(data_frame , test_size=0.2 , random_state=42)
train_set
```

```python
#splitting features and labels
features = train_set.drop("median_house_value" , axis=1)
labels = train_set["median_house_value"].copy()
#ENCODING CATEGORICAL FEATURES 
from sklearn.preprocessing import OneHotEncoder
cat_encoder = OneHotEncoder()
housing_cat = features[["ocean_proximity"]]
housing_cat_one_hot = cat_encoder.fit_transform(housing_cat)
housing_cat_one_hot_dataframe = pd.DataFrame(
    housing_cat_one_hot.toarray(),
    columns = cat_encoder.get_feature_names_out(),
    index = features.index 
)
#lets drop ocean proximity form the features dataset
features.drop("ocean_proximity" , axis = 1 , inplace = True)
#now concating into feature dataframe with the help of the  pandas
import pandas as pd
features_as_x  = pd.concat([features , housing_cat_one_hot_dataframe] , axis = 1)
features_as_x


```

```python
#now using simple imputer to fill missing values with mediaN 
from sklearn.impute import SimpleImputer
imputer  = SimpleImputer(strategy="median")
X = imputer.fit_transform(features_as_x)
#making changes into orginal features dataframe
features_as_x = pd.DataFrame(X , columns=features_as_x.columns , index = features_as_x.index)
```

```python
#feature scaling why feature scaling is important ? = 
# to ensure that the all features are comparable 
# many ml models think all features are comparable and on a same scale
# large features can dominate smaller which cna affect the output (label) or performance of model 
```

```python
#using min - max scaling also known as normalization
from sklearn.preprocessing import MinMaxScaler
feature_range = (-1 , 1)
min_max_scaler_model = MinMaxScaler(feature_range=feature_range)
data_scaled_with_normalizaton = min_max_scaler_model.fit_transform(features_as_x)
data_scaled_with_normalizaton_dataframe = pd.DataFrame(data_scaled_with_normalizaton , columns=features_as_x.columns , index=features_as_x.index)
data_scaled_with_normalizaton_dataframe
```

```python
#Standardiztiaon also known as z score normalization it is the feature scaling method used to scale features on the
#same scale so the result or outcome have mean of 0 and  standard deviation of 1 
#formula = z score normalization = (X - mean) / standard deviation
#now lets see the practical example of standardization using the sckit learn
from sklearn.preprocessing import StandardScaler
#first step done
model = StandardScaler()
#second step done
data_scaled_with_standardization = model.fit_transform(features_as_x)
data_scaled_with_z_score_dataframe = pd.DataFrame(data_scaled_with_standardization ,
                                                  columns = features_as_x.columns,
                                                  index = features_as_x.index)
data_scaled_with_z_score_dataframe
```
