```python
import pandas as pd
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score
import matplotlib.pyplot as plt


df = pd.read_csv("housing.csv")


df.fillna(0, inplace=True)

df = pd.get_dummies(df, drop_first=True)

Label_as_y = df["median_house_value"]
Features_as_x = df.drop("median_house_value", axis=1)


model = LinearRegression()
model.fit(Features_as_x, Label_as_y)


first_row = Features_as_x.iloc[0:1]  
prediction = model.predict(first_row)
print("Predicted house value:", prediction[0])



predictions = model.predict(Features_as_x)
print("Mean Squared Error:", mean_squared_error(Label_as_y, predictions))
print("RÂ² Score:", r2_score(Label_as_y, predictions))


plt.scatter(Label_as_y, predictions, alpha=0.5)
plt.xlabel("Actual House Value")
plt.ylabel("Predicted House Value")
plt.title("Actual vs Predicted House Prices")
plt.show()
```
